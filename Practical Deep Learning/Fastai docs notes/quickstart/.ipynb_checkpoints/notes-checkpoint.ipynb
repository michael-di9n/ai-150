# Computer vision
e.g., recognizing images of cats and dogs from 37 diffrent breeds
This one uses a pretrained model on 1.3mill images and then fine-tuned using transfer learning 

```python


# Oxford IIIT Pet Dataset
path = untar_data(URLs.PETS)/'images'

def is_cat(x): return x[0].isupper()
dls = ImageDataLoaders.from_name_func(
    path, get_image_files(path), valid_pct=0.2, seed=42,
    label_func=is_cat, item_tfms=Resize(224))

learn = vision_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(1)
```

```python
# Make sure the current working directory is ..quickstart
import os
os.getcwd()
```

```python
# We can do inference with the predict mathod
img = PILImage.create('./quickstart/images/puppy.jpg')
img
```

```python
is_cat,_,probs = learn.predict(img)
print(f"Is this a cat?: {is_cat}.")
print(f"Probability it's a cat: {probs[1].item():.6f}")
```

# Computer vision segmentation
Image segmentation is a computer vision technique to partition a digital image into discrete groups of pixels
- Using specific shaped segments

```python
# This one is based of the Camvid dataset for Car camera
path = untar_data(URLs.CAMVID_TINY)
dls = SegmentationDataLoaders.from_label_func(
    path, bs=8, fnames = get_image_files(path/"images"),
    label_func = lambda o: path/'labels'/f'{o.stem}_P{o.suffix}',
    codes = np.loadtxt(path/'codes.txt', dtype=str)
)

learn = unet_learner(dls, resnet34)
learn.fine_tune(8)
```

```python
# Use show results to view how the model performed
learn.show_results(max_n=4, figsize=(7,8))
```

```python
# Plotting the k instances that contributes mode to the validation loss using SegmentationInterpetatio
interp = SegmentationInterpretation.from_learner(learn)
interp.plot_top_losses(k=2)

# Note: loss is the measure of inaccuracy of a predition. The lower the loss the better the higher the worse
# Generally Validation loss and training loss should be similar - if the training loss is much lower than validation then 
# we are overfitting on the training data
```

Examining the high loss images the result is probably due to the poor lighting

# Natural Language Processing
Classify the sentiment of a movie review etc

```python
dls = TextDataLoaders.from_folder(untar_data(URLs.IMDB), valid='test')
learn = text_classifier_learner(dls, AWD_LSTM, drop_mult=0.5, metrics=accuracy)
learn.fine_tune(2, 1e-2)
```

```python
learn.predict("I really liked that movie!")
```

# Tabular 
Building models from plain tabular data (exteremely common) uses same basic techiniques as above


Here the model predicts whether a person is a high-income earner based on socioeconomic background

```python
path = untar_data(URLs.ADULT_SAMPLE)

dls = TabularDataLoaders.from_csv(path/'adult.csv', path=path, y_names="salary",
    cat_names = ['workclass', 'education', 'marital-status', 'occupation',
                 'relationship', 'race'],
    cont_names = ['age', 'fnlwgt', 'education-num'],
    procs = [Categorify, FillMissing, Normalize])

learn = tabular_learner(dls, metrics=accuracy)
learn.fit_one_cycle(2)
```

# Reccomendation system
Very important in e-commerce and applications
This one uses a collabmodel to train

```python
path = untar_data(URLs.ML_SAMPLE)
dls = CollabDataLoaders.from_csv(path/'ratings.csv')
learn = collab_learner(dls, y_range=(0.5,5.5))
learn.fine_tune(6)
```
